## Row picture
세 개의 선형 방정식들로 구성된 연립방정식을 row picture로 접근한다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 방정식을 행렬 형태로 표현
A = np.array([
    [1, 2, 3],
    [2, 5, 2],
    [6, -3, 1]
])

B = np.array([6, 4, 2])

# 방정식의 해를 구함
solution = np.linalg.solve(A, B)

# 그래프를 그리기 위한 3D 플롯 생성
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# 각 방정식에 대한 그래프 생성
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# 첫 번째 방정식: x + 2y + 3z = 6
Z1 = (6 - X - 2*Y) / 3.0
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100)

# 두 번째 방정식: 2x + 5y + 2z = 4
Z2 = (4 - 2*X - 5*Y) / 2.0
ax.plot_surface(X, Y, Z2, alpha=0.5, facecolors='g', rstride=100, cstride=100)

# 세 번째 방정식: 6x - 3y + z = 2
Z3 = 2 - 6*X + 3*Y
ax.plot_surface(X, Y, Z3, alpha=0.5, facecolors='r', rstride=100, cstride=100)

# 해 (교점)를 그래프에 표시
ax.scatter(solution[0], solution[1], solution[2], color='k', s=100, label='Intersection point', marker='o')

# 그래프 설정
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Three Plane Intersection')
ax.legend()

plt.show()
```

## Column Picture

```python
import numpy as np
import matplotlib.pyplot as plt

# 주어진 벡터
a = np.array([1, 2, 6])
b = np.array([2, 5, -3])
c = np.array([3, 2, 1])
target = np.array([6, 4, 2])

# 선형 조합을 통해 target 벡터를 만들기 위한 계수 행렬
matrix = np.vstack([a, b, c]).T

# 선형 방정식 풀이
coefficients = np.linalg.lstsq(matrix, target, rcond=None)[0]

print(coefficients)

# 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='b', label='a')
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='g', label='b')
ax.quiver(0, 0, 0, c[0], c[1], c[2], color='r', label='c')
ax.quiver(0, 0, 0, target[0], target[1], target[2], color='k', label='Target')
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.legend()
plt.show()
```
