# 1.Vectors

![img](./imgs/0001.png)


- n개의 숫자들이 모여 n차원 벡터를 만들 수 있다.
- 벡터가 가진 원소(component)의 수가 곧 벡터가 속한 차원(dimension)을 나타낸다.
- 일반적으로 벡터의 기본 표기법은 column vector이며, row vector는 column vector의 transpose이다.

벡터는 그래프상에서 화살표로 나타낼 수 있으며 이는 방향과 크기를 동시에 나타내고 있음을 의미한다.  
벡터를 그리기 위해서는 시작점과 종점을 이용해야 한다. 시작점은 항상 원점(0, 0)이며 종점은 벡터 자신 $v_1, v_2$에 해당한다.

```python
import numpy as np
import matplotlib.pyplot as plt

start_point = np.array([0, 0])
end_point = np.array([3, 4])
vector = end_point - start_point

## 0. figure 생성
plt.figure()

## 1. 좌표축 나타내기
plt.axvline(x=0, color='gray', linewidth=2)  # x축
plt.axhline(y=0, color='gray', linewidth=2)  # y축

## 2. 각각의 축에서 그릴 수 있는 범위
plt.xlim(-9, 10)
plt.ylim(-9, 10)

## 3. 격자모양 그리기
plt.grid()

## 4. x축과 y축의 눈금 단위를 1로 설정.
plt.xticks(np.arange(-9, 10, step=1))
plt.yticks(np.arange(-9, 10, step=1))

plt.quiver(*start_point, *vector, angles='xy', scale_units='xy', scale=1, color='blue')

## optional. 각 축과 그래프의 제목을 지정
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Example')

## 5. 그래프를 표시.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

v1 = np.array([0, 0, 9])
v2 = np.array([4, 5, 6])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', label='Vector')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', label='Vector')

# Set axis limits
ax.set_xlim(-10, 10)  # Adjust the limits according to your vector
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Vector')

# Show the plot
plt.legend()
plt.show()
```

벡터의 덧셈과 뺄셈은
- 두 벡터 간의 차원이 같을 때 즉, component의 수가 일치할 때 성립한다.
- 동일한 위치에 있는 component끼리 서로 더하거나 뺀다.

```python
import numpy as np

v = np.array([1, 0])
w = np.array([3, 4])

vec_add = v + w
vec_sub = v - w

print(vec_add)
print(vec_sub)
```

스칼라 곱(scalar multiplication)은 벡터에 상수배를 해주는 것으로 벡터의 원소들에 임의의 상수 c가 곱해진다.

```python
v = np.array([3, 7])

vec_scalar_mul = 2 * v
print(vec_scalar_mul)
```

# 2.Linear Combinations

![imgs](./imgs/0004.png)

위에서 등장한 Vector Addition과 Scalar multiplication이 선형대수의 핵심적인 연산이다.  
"Lininear algebra is built on these operations $v + w$ and $cv$ and $dw$ -- adding vectors and multiplying by scalars."

- 벡터의 덧셈연산과 스칼라배 연산을 결합(combine)하게 되면 $cv + dw$로 나타낼 수 있게 되는데, 이를 Linear Combination이라고 한다.
- Linear Combination에 참여한 벡터들의 가능한 모든 조합을 모아보면 특정한 공간(space)를 구성하게 된다.
- 이 공간에는 항상 영벡터(zero vector)가 포함되어야만 한다. 즉, 영벡터가 포함되지 않는 경우 공간은 정의되지 않는다.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the vector v
v = np.array([1, 2, 3])

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original vector v as a point
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='Vector v')

# Plot the linear combinations for different values of c
for c in range(11):
    cv = c * v
    ax.quiver(0, 0, 0, cv[0], cv[1], cv[2], color='g', alpha=0.3)


# Set integer ticks and labels
ax.set_xticks(np.arange(0, 10))
ax.set_yticks(np.arange(0, 10))
ax.set_zticks(np.arange(0, 10))
ax.set_xticklabels(map(str, np.arange(0, 10)))
ax.set_yticklabels(map(str, np.arange(0, 10)))
ax.set_zticklabels(map(str, np.arange(0, 10)))

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Linear Combinations of Vector v')

# Show the plot
plt.legend()
plt.show()
```
