```python
import numpy as np
import matplotlib.pyplot as plt

# Given vectors
v = np.array([1, 2])
v1 = np.array([1, 0])
v2 = np.array([0, 2])

# Calculate the lengths of the vectors
length_v = np.sqrt(np.sum(v ** 2))
length_v1 = np.sqrt(np.sum(v1 ** 2))
length_v2 = np.sqrt(np.sum(v2 ** 2))

# Create a plot to visualize the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector v (1, 2)')
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector (1, 0)')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector (0, 2)')
plt.xlim(-3, 4)
plt.ylim(-3, 4)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Vectors and Their Lengths")
plt.legend()
plt.show()

# Print the calculated lengths
print("Length of the vector v:", length_v)
print("Length of the vector (1, 0):", length_v1)
print("Length of the vector (0, 2):", length_v2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Standard unit vectors
i_hat = np.array([1, 0])
j_hat = np.array([0, 1])

# Create a plot to visualize the standard unit vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, i_hat[0], i_hat[1], angles='xy', scale_units='xy', scale=1, color='red', label='i-hat')
plt.quiver(0, 0, j_hat[0], j_hat[1], angles='xy', scale_units='xy', scale=1, color='blue', label='j-hat')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Standard Unit Vectors in xy Plane")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Two perpendicular vectors
v1 = np.array([2, 0])  # Horizontal vector
v2 = np.array([0, 3])  # Vertical vector

# Create a plot to visualize the vectors and the right angle triangle
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector v2')
plt.xlim(-2, 4)
plt.ylim(-2, 4)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Perpendicular Vectors and Right Angle Triangle")
plt.legend()

# Adding the perpendicular lines to form the right angle triangle
plt.plot([v1[0], v1[0]], [v1[1], v1[1] + v2[1]], color='green', linestyle='--')
plt.plot([v2[0], v1[0]], [v2[1], v2[1]], color='green', linestyle='--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two vectors v and w
v = np.array([2, 0])   # Horizontal vector
w = np.array([0, 3])   # Vertical vector

# Calculate the difference vector between v and w
difference_vector = v - w

# Calculate the length of the hypotenuse (v-w)
hypotenuse_length = np.linalg.norm(difference_vector)

# Create a plot to visualize the vectors and the right angle triangle
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector w')
plt.quiver(w[0], w[1], difference_vector[0], difference_vector[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector v-w')
plt.xlim(-2, 4)
plt.ylim(-2, 4)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Perpendicular Vectors and Right Angle Triangle")
plt.legend()

# Adding the hypotenuse line (v-w)
plt.plot([w[0], v[0]], [w[1], v[1]], color='purple', linestyle='--', label='Hypotenuse (v-w)')
plt.annotate(f'Length: {hypotenuse_length:.2f}', (0.5, 1), color='purple')

plt.legend()
plt.show()
```
