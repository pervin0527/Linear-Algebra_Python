## 0. Row picture & Column picture

```python
import numpy as np
import matplotlib.pyplot as plt

# 방정식의 계수
a1, b1, c1 = 2, -1, 0
a2, b2, c2 = -1, 2, 0

# 연립방정식의 해 구하기
A = np.array([[a1, b1], [a2, b2]])
B = np.array([c1, c2])
solution = np.linalg.solve(A, B)

# x 범위 설정
x_vals = np.linspace(-5, 5, 400)

# y 값 계산
y_vals1 = (c1 - a1 * x_vals) / b1
y_vals2 = (c2 - a2 * x_vals) / b2

# 교점 좌표
intersection_x = solution[0]
intersection_y = solution[1]

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals1, label='2x - y = 0')
plt.plot(x_vals, y_vals2, label='-x + 2y = 0')
plt.scatter(intersection_x, intersection_y, color='red', label='Intersection')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of Linear Equations with Intersection')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

print("Intersection point:", solution)
```

```python
## column picture

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 주어진 벡터
vectors = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])

# 원점으로 가는 선을 그리기 위한 점들
t = np.linspace(-1, 1, 10)
origin_line = np.array([t*0, t*0, t*0])

# 그래프 그리기
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 벡터들 그리기
for vector in vectors:
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color='b', arrow_length_ratio=0.1)

# 원점으로 가는 선 그리기
ax.plot(origin_line[0], origin_line[1], origin_line[2], color='r', linestyle='--')

# 축 설정
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Linear Combination of Vectors')

plt.show()
```

## 1.Linearly Independent

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 첫 번째 벡터 u
u = np.array([1, -1, 0])

# 두 번째 벡터 v
v = np.array([0, 1, -1])

# 가중치 범위 설정
weights = np.linspace(-2, 2, 100)

# 선형 결합된 벡터들을 계산
linear_combinations = [w * u for w in weights]

# 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 선형 결합된 벡터들을 그래프에 표시
for lc in linear_combinations:
    ax.quiver(0, 0, 0, lc[0], lc[1], lc[2], color='b', alpha=0.5)

# 벡터 v를 그래프에 표시
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='g', label='v')

# 축 범위 설정
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# 축 레이블
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 벡터 설명을 나타낼 범례 추가
ax.legend()

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 주어진 벡터들
u = np.array([1, -1, 0])
v = np.array([0, 1, -1])
w = np.array([0, 0, 1])

# 벡터들을 3D 그래프에 표시
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 벡터들을 원점에서 시작하도록 그리기
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u=[1, -1, 0]')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='g', label='v=[0, 1, -1]')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='b', label='w=[0, 0, 1]')

# 벡터들이 만드는 평면을 그리기 위한 점들
plane_points = np.array([u, v, u + v])  # u, v, u+v를 이용하여 평면을 생성

# 평면을 시각화하기 위한 코드
normal = np.cross(u, v)  # 평면의 법선 벡터 계산
d = -np.dot(normal, u)  # 평면의 방정식 계산을 위한 값

xx, yy = np.meshgrid(range(-2, 2), range(-2, 2))
zz = (-normal[0] * xx - normal[1] * yy - d) * 1.0 / normal[2]
ax.plot_surface(xx, yy, zz, color='gray', alpha=0.5)

# 벡터 w가 평면 내에 있는지 확인
w_projection = np.dot(w, normal) * normal
ax.quiver(0, 0, 0, w_projection[0], w_projection[1], w_projection[2], color='m', label='w_proj')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.legend()

plt.show()
```
