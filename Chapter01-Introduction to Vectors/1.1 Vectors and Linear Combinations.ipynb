## 2D Vector

```python
import numpy as np
import matplotlib.pyplot as plt

# 2차원 벡터의 시작점과 끝점 좌표
start_point = np.array([0, 0])
end_point = np.array([3, 4])

# 벡터 그리기
plt.figure()
plt.quiver(*start_point, *(end_point - start_point), angles='xy', scale_units='xy', scale=1, color='b')

# x축과 y축 굵게 그리기
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

# x축과 y축 간격 설정
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(-10, 11, 1))

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Vector')

plt.grid()
plt.show()
```

## 3D Vector

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3차원 벡터의 시작점과 끝점 좌표
start_point = np.array([0, 0, 0])
end_point = np.array([1, 2, 3])

# 그래프 생성
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# 벡터 그리기
ax.quiver(*start_point, *(end_point - start_point), color='b')

# x, y, z 축 범위 설정
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 4])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.title('3D Vector')
plt.show()
```

## Vector Addition(2D)

```python
import numpy as np
import matplotlib.pyplot as plt

# 두 개의 2차원 벡터 정의
vector1 = np.array([2, 3])
vector2 = np.array([-1, 4])

# 벡터 덧셈
result_vector = vector1 + vector2

# 그래프 생성
plt.figure()

# 원점에서 vector1까지의 벡터 그리기
plt.quiver(0, 0, *vector1, angles='xy', scale_units='xy', scale=1, color='b', label='Vector 1')

# 원점에서 vector2까지의 벡터 그리기
plt.quiver(0, 0, *vector2, angles='xy', scale_units='xy', scale=1, color='g', label='Vector 2')

# 원점에서 결과 벡터까지의 벡터 그리기
plt.quiver(0, 0, *result_vector, angles='xy', scale_units='xy', scale=1, color='r', label='Vector Sum')

plt.xlim(-10, 11)
plt.ylim(-10, 11)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Addition')
plt.legend()
plt.grid()
plt.show()
```

## Scalar Multiplication(2D)

```python
import numpy as np
import matplotlib.pyplot as plt

# 2차원 벡터 정의
vector = np.array([2, 3])

# 스칼라곱 적용
scalar = 2
scaled_vector = scalar * vector

# 그래프 생성
plt.figure()

# 원점에서 원래 벡터까지의 벡터 그리기
plt.quiver(0, 0, *vector, angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')

# 원점에서 스칼라곱 적용된 벡터까지의 벡터 그리기
plt.quiver(0, 0, *scaled_vector, angles='xy', scale_units='xy', scale=1, color='r', label='Scaled Vector')

# 굵은 x축과 y축 그리기
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

# x, y 축 간격 설정
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(-10, 11, 1))

plt.xlim(-10, 11)
plt.ylim(-10, 11)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scalar Multiplication of a Vector')
plt.legend()
plt.grid()
plt.show()
```

## Linear Combination(3D, Only One Vector.)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 3차원 벡터 정의
u = np.array([1, 2, 3])

# 스칼라 값 범위 설정
scalar_range = np.linspace(-2, 2, 20)

# 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 원점에서 3차원 벡터 u까지의 벡터 그리기
ax.quiver(0, 0, 0, *u, color='b', label='Vector u')

# Linear combination 결과 벡터들 그리기
for scalar in scalar_range:
    scaled_vector = scalar * u
    ax.quiver(0, 0, 0, *scaled_vector, color='r', alpha=0.2)

# 축 범위 설정
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.title('Linear Combination of a 3D Vector')
plt.legend()
plt.show()
```

## Linear Combination(3D, Two vectors)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 3차원 벡터 정의
u = np.array([1, 2, 3])
v = np.array([-1, 1, 2])

# 스칼라 값 범위 설정
scalar_range = np.linspace(-2, 2, 20)

# 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 원점에서 벡터 u와 v까지의 벡터 그리기
ax.quiver(0, 0, 0, *u, color='b', label='Vector u')
ax.quiver(0, 0, 0, *v, color='g', label='Vector v')

# 모든 Linear combination 결과 벡터들 그리기
for scalar_u in scalar_range:
    for scalar_v in scalar_range:
        combined_vector = scalar_u * u + scalar_v * v
        ax.quiver(0, 0, 0, *combined_vector, color='r', alpha=0.2)

# 축 범위 설정
ax.set_xlim([-7, 8])
ax.set_ylim([-7, 8])
ax.set_zlim([-7, 8])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.title('Linear Combinations of Vectors u and v')
plt.legend()
plt.show()
```

## Linear Combinations(3D, three vectors, )

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from itertools import product

# 3차원 벡터 정의 (선형 독립적인 벡터들)
vectors = [np.array([1, 0, 0]), np.array([0, 1, 0]), np.array([0, 0, 1])]

# 스칼라 값 범위 설정
scalar_range = np.linspace(-2, 2, 20)
print(scalar_range)

# 모든 가능한 스칼라 값 조합 계산
scalar_combinations = list(product(scalar_range, repeat=len(vectors)))

# 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 원점에서 벡터들까지의 벡터 그리기
for vec in vectors:
    ax.quiver(0, 0, 0, *vec, color=np.random.rand(3), label=f'Vector {vec}')

# 모든 Linear combination 결과 벡터들 그리기
for scalar_comb in scalar_combinations:
    combined_vector = np.sum(scalar * vec for scalar, vec in zip(scalar_comb, vectors))
    ax.quiver(0, 0, 0, *combined_vector, color='r', alpha=0.2)

# 축 범위 설정
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.title('Linear Combinations of Three Linearly Independent Vectors')
plt.legend()
plt.show()
```
